```python
sumNegL1=0
sumNegL2=0
sumNegL3=0

sumPosL1=0
sumPosL2=0
sumPosL3=0

for i in range (1, 51):
    l1=ImaxDemanda[f"  I H {i} L1 prom. 10 min [A]  "].iloc[0]
    l2=ImaxDemanda[f"  I H {i} L2 prom. 10 min [A]  "].iloc[0]
    l3=ImaxDemanda[f"  I H {i} L3 prom. 10 min [A]  "].iloc[0]
    

    P1=pn.loc[idMax][f"P{i} L1 prom. 10 min \n[Watts]  "]
    P2=pn.loc[idMax][f"P{i} L2 prom. 10 min \n[Watts]  "]
    P3=pn.loc[idMax][f"P{i} L3 prom. 10 min \n[Watts]  "]

    if P1<0:
        sumNegL1+=l1**2
    else:
        sumPosL1+=l1**2

    if P2<0:
        sumNegL2+=l2**2
    else:
        sumPosL2+=l2**2
    
    if P3<0:
        sumNegL3+=l3**2
    else:
        sumPosL3+=l3**2
```

```python
ImaxDemanda["HGL1"]=np.sqrt(sumNegL1)/np.sqrt(sumPosL1)
ImaxDemanda["HGL2"]=np.sqrt(sumNegL2)/np.sqrt(sumPosL2)
ImaxDemanda["HGL3"]=np.sqrt(sumNegL3)/np.sqrt(sumPosL3)
ImaxDemanda[["HGL1","HGL2","HGL3"]]
```

Considerando que Isc entre IL ES 2.26, debemos usar los siguientes criterios:



```python
IEE=excel["IEEE-519 2022"].copy().iloc[:-2].set_index(['Fecha        ', ' Hora'])
IEE=IEE[IEE.columns[4:]]*100
```

```python
IEE
```

```python
def colorear(val, limite):
    color = 'background-color: #f5b042' if val >= limite else ''
    return color

for i in range(2,51): #Armonicas de la 2 a la 50....
    ss=[f' I H {i} L1 prom. 10 min \n[% de IL Dem]', f' I H {i} L2 prom. 10 min \n[% de IL Dem]', f' I H {i} L3 prom. 10 min \n[% de IL Dem]'] #Subset
    if i ==2:
        IEE_styled = IEE.style.map(lambda val: colorear(val, 2),        subset=ss)
        continue

    if 3 <= i <7: #Si está entre 2-6

        if i%2==1: #par
            IEE_styled = IEE_styled.map(lambda val: colorear(val, 2.0), subset=ss)
            continue
        else:      #impar
            IEE_styled = IEE_styled.map(lambda val: colorear(val, 4.0), subset=ss)
            continue

    elif 7<=i<11: #Si está entre 7-10
        IEE_styled = IEE_styled.map(lambda val: colorear(val, 4.0), subset=ss)
        continue
    
    elif 11<=i<17: #Si está entre 11-16
        IEE_styled = IEE_styled.map(lambda val: colorear(val, 2.0), subset=ss)
        continue

    elif 17<=i<23: #Si está entre 17-22
        IEE_styled = IEE_styled.map(lambda val: colorear(val, 1.5), subset=ss)
        continue

    elif 23<=i<35: #Si está entre 23-34
        IEE_styled = IEE_styled.map(lambda val: colorear(val, 0.6), subset=ss)
        pass

    else:           #Si está entre 35-50
        IEE_styled = IEE_styled.map(lambda val: colorear(val, 0.3), subset=ss)
        continue
```

```python
violaciones = 0

for i in range(2,50):
    IEE_arm=IEE[[f' I H {i} L1 prom. 10 min \n[% de IL Dem]', f' I H {i} L2 prom. 10 min \n[% de IL Dem]', f' I H {i} L3 prom. 10 min \n[% de IL Dem]']]
    lim=lim_Impares if (i%2 == 1) else lim_Pares

    if i <11:
        lim=lim[0]
    elif i<17:
        lim=lim[1]
    elif i<23:
        lim=lim[2]
    else:
        lim=lim[3]
    
    #Contar valores que exceden el límite en las armonicas
    violaciones = violaciones+IEE_arm.apply(lambda row: (row >= lim).sum(), axis=1)


violaciones
```
