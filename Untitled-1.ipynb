```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
global Vm, Vd, R
Vm  = 170
R   = 0.24
Vd =  72

def I(a):
    Vs=(2*Vm/np.pi)*np.cos(a)
    V=Vs-Vd
    return V / R
```

```python
Imax=((2*Vm/np.pi)-Vd)/R
Imin=-1*((2*Vm/np.pi)+Vd)/R
x0=-Vd/R
```

```python
Imax
```

```python
l=np.linspace(0, 2*np.pi, 1005)
plt.figure(figsize=(11, 5))  # Ancho = 10 pulgadas, Alto = 5 pulgadas

y=I(l)
plt.plot(l, y, label="I")
plt.plot(l, [Imax]*len(l), "--",label=f"I max=[(2 Vm/pi)+Vd]/R   ={round(Imax,3)}")
plt.plot(l, [Imin]*len(l), "--",label=f"I min=-1[(2 Vm/pi)+Vd]/R = {round(Imin,3)}")
plt.plot(l, [x0]*len(l), "-.", label=f"- Vd/R={x0}")
plt.plot(l, [0]*len(l), "-k", label="y=0")



plt.grid()
plt.legend()
plt.xlabel("Alpha (a) en radianes")
plt.ylabel(" Corriente (I) ")
plt.title("Corriente como funcin de alpha [ I = ( (2*Vm/pi)*cos(a)-Vd)/R ]", fontweight='bold')
plt.xticks([0.843, np.pi/2,np.pi,3*np.pi/2, 5.44, 2*np.pi])
print("Vd=", Vd)
print("Vm=",Vm)
print("R=", R)
```
